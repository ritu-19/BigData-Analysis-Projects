Now, calculate the mean for 'meantempi' for the days that are Saturdays or Sundays (weekend):

````python
import datetime as dt
#import Series.dt.weekday
def avg_weekend_temperature(df):
    
    df['DATEn'] = pandas.to_datetime(df['DATEn'], errors='coerce')
    df['weekday'] = df['DATEn'].dt.dayofweek
    df_new = df.loc[(df['weekday'] == 5) | (df['weekday'] ==6)]
    mean_temp_weekends = df_new['meantempi'].mean()
    return mean_temp_weekends

mean_temp_weekends = avg_weekend_temperature(weather_data)
print("Mean for 'meantempi for weekends: ",mean_temp_weekends)
#weather_data.head()
````
